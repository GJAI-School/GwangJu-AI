```python
import numpy as np

def compute_a(x, y, mean_x, mean_y):
    dividend = 0
    
    for i in range(len(x)):
        dividend += (x[i] - mean_x) * (y[i]-mean_y)
    
    divisor = sum([(x1 - mean_x)**2 for x1 in x])
    a = dividend / divisor
    return a

x = [2, 4, 6, 8]
y = [81, 93, 91, 97]

mx = np.mean(x)
my = np.mean(y)

a = compute_a(x, y, mx, my)
b = my - (mx * a)
print(a, b)

y1 = [a * x1 + b for x1 in x]

plt.grid()
plt.plot(x, y, 'bo')
plt.plot(x, y1, 'r-o')
plt.show()
```

```python
def MSE(y_pred, y):
    return np.sum((y_pred - y)**2) / len(y)
```

```python
import matplotlib.pyplot as plt
import numpy as np

n_samples = 200
x1 = np.random.randn(n_samples)
y = 2 * x1 + 4 + np.random.randn(n_samples)

num_epochs = 10
lr = 0.7

w1 = np.random.uniform()
b = np.random.uniform()

for epoch in range(num_epochs):
    y_pred = w1 * x1 + b
    cost = MSE(y_pred, y)
    
    w1 = w1 - lr * ((y_pred - y) * x1).mean()
    b = b - lr * ((y_pred - y).mean())
    print('w:', w1, 'b:', b, 'cost:',cost)
    if epoch == num_epochs:
        break
    plt.plot(x1, y_pred, color='green')

plt.plot(x1, y_pred, color='red')
plt.plot(x1, y, 'bo')
plt.show()
```
