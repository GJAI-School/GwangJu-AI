```python
import pandas as pd
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv("./data/bmi_500.csv")
```

```python
df
```

```python
df[df.Gender == 'Male'] 
```

```python
t = df.shape[0] * 0.8
```

```python
train_x = df.loc[:t+1,'Gender':'Weight']
train_y = df.loc[:t+1, 'Label']

test_x = df.loc[t+1:,'Gender':'Weight']
test_y = df.loc[t+1:, 'Label']
```

```python
train_x
```

```python
model = KNeighborsClassifier(n_neighbors=2)
model.fit(train_x, train_y)

metrics.accuracy_score(test_x, test_y)
```
